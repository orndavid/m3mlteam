## Results
We forgot that regression does not give a 0-1 percentage R value. We need an alternative method for computing the errors

```python
df_p = df_r
df_p.set_index("percentage")
df_p = (df_p - df_p.min())/(df_p.max() - df_p.min())
```

```python
ax = df_p.plot()
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_ylabel("Normalized RMS")
```
